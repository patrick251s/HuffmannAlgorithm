```python
import matplotlib.pyplot as plt
import networkx as nx
import math

string = input("Podaj tekst do zakodowania (spacja jako _): ")

class NodeTree(object):

    def __init__(self, left=None, right=None):
        self.left = left
        self.right = right

    def children(self):
        return (self.left, self.right)

    def nodes(self):
        return (self.left, self.right)

    def __str__(self):
        return '%s_%s' % (self.left, self.right)

def huffman_code_tree(node, left=True, binString=''):
    if type(node) is str:
        return {node: binString}
    (l, r) = node.children()
    d = dict()
    d.update(huffman_code_tree(l, True, binString + '0'))
    d.update(huffman_code_tree(r, False, binString + '1'))
    return d

def get_characteristic(huffmanCode, probabilities):
  length_of_code = 0
  mean_length = 0
  entropy = 0
  for i in range(len(huffmanCode)):
    length_of_code += len(huffmanCode[i])
    mean_length += len(huffmanCode[i]) * probabilities[i]
    entropy += probabilities[i]*math.log2(1/probabilities[i])

  print("Dlugość kodu = ", length_of_code)
  print("Średnia długość słowa kodowego = ", round(mean_length, 4))
  print("Entropia = ", round(entropy, 4))

def getParentNode(node):
  if(node == "0" or node == "1"):
    return ""
  else:
    return node[:-1]

def countNodePosition(node):
  if(node == ""):
    return (10000, 10000)
  horizontal = 10000 - 1000*len(node)
  vertical = 10000
  for i in range(len(node)):
    sumElement = 10000
    for j in range(i+1):
      sumElement = sumElement/2
    if(node[i] == "0"):
      vertical -= sumElement
    elif(node[i] == "1"):
      vertical += sumElement
  return (vertical, horizontal)

def draw_graph(chars, huffmanCode):
  theLongestSize = len(max(huffmanCode, key = len))

  for s in range(theLongestSize):
    for i in range(len(huffmanCode)):
      parent = getParentNode(huffmanCode[i])
      if parent not in huffmanCode:
        huffmanCode.append(parent)

  # Node: Parent
  parents = {}
  for i in range(len(huffmanCode)):
    parents[huffmanCode[i]] = getParentNode(huffmanCode[i])

  # Separate Nodes
  nodesList = list(parents.keys())
  parentsList = list(parents.values())
  elabels = {}

  plt.figure(1,figsize=(200,80))
  G = nx.Graph()
  for n in range(len(nodesList)):
    if(nodesList[n] == "" and parentsList[n] == ""):
        continue
    G.add_edge(nodesList[n], parentsList[n])
    elabels[(nodesList[n], parentsList[n])] = nodesList[n][-1]

  position = {}
  for i in range(len(huffmanCode)):
    position[huffmanCode[i]] = countNodePosition(huffmanCode[i])

  labeldict = {}
  for i in range(len(chars)):
    labeldict[huffmanCode[i]] = chars[i]

  nx.draw_networkx(G, pos = position, labels = labeldict, font_size = 26, node_size = 2000, node_color = "white", edgecolors = "black", width=5)
  nx.draw_networkx_edge_labels(G, pos = position, font_size=26,edge_labels = elabels, font_color = "red")

  plt.margins(x=0, y=0)
  plt.axis("off")
  plt.show()


freq = {}
for c in string:
    if c in freq:
        freq[c] += 1
    else:
        freq[c] = 1

freq = sorted(freq.items(), key=lambda x: x[1], reverse=True)
nodes = freq

probabilities = []
for i in range(len(freq)):
  probabilities.append(freq[i][1]/len(string))

while len(nodes) > 1:
    (key1, c1) = nodes[-1]
    (key2, c2) = nodes[-2]
    nodes = nodes[:-2]
    node = NodeTree(key1, key2)
    nodes.append((node, c1 + c2))
    nodes = sorted(nodes, key=lambda x: x[1], reverse=True)

huffmanCode = huffman_code_tree(nodes[0][0])
sortedHuffmanCode = []

print(' Char | Huffman code | Frequency | Probabilities')
print('-----------------------------------------------------')
for a in range(len(freq)):
    sortedHuffmanCode.append(huffmanCode[freq[a][0]])
    print(' %-4r |%12s |%12s |%12s' % (freq[a][0], huffmanCode[freq[a][0]], freq[a][1], round(probabilities[a],4)))
print('-----------------------------------------------------')

get_characteristic(sortedHuffmanCode, probabilities)

chars = []
hCodes = []
for i in range(len(freq)):
  chars.append(freq[i][0])
  hCodes.append(huffmanCode[freq[i][0]])

draw_graph(chars, hCodes)
```
